```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('CSV files/spotify_dataset.csv')
```

```python
df
```

```python
df_drop = df.drop(['Index', 'Highest Charting Position', 'Number of Times Charted', 'Song ID', 'Duration (ms)', 'Chord', 'Weeks Charted'], axis=1)
```

```python
df_drop
```

```python
df_drop.isnull().sum()
```

```python
df_drop.dtypes
```

```python
df_drop['Week of Highest Charting']
```

```python
df_drop[['First Week Highest Charting', 'Second Week of Highest Charting']] = df_drop['Week of Highest Charting'].str.split('--', 1, expand=True)
```

```python
df_drop.drop(['Week of Highest Charting'], axis=1)
```

```python
df_drop['First Week Highest Charting'] = pd.to_datetime(df_drop['First Week Highest Charting'], format='%Y-%m-%d')
df_drop['Second Week of Highest Charting'] = pd.to_datetime(df_drop['Second Week of Highest Charting'], format='%Y-%m-%d')
```

```python
df_drop.drop(['Release Date'],axis=1)
```

```python
df_drop['Day of First Highest Charting'] = df_drop['First Week Highest Charting'].dt.day_name()
df_drop['Day of Second Highest Charting'] = df_drop['Second Week of Highest Charting'].dt.day_name()
```

```python
df_drop['Month of First Highest Charting'] = pd.DatetimeIndex(df_drop['First Week Highest Charting']).month
df_drop['Month of Second Highest Charting'] = pd.DatetimeIndex(df_drop['Second Week of Highest Charting']).month
```

```python
df_drop.head()
```

```python
df_drop.drop(['Week of Highest Charting', 'First Week Highest Charting', 'Second Week of Highest Charting'], axis=1, inplace=True)
```

```python
df_drop.drop(['Release Date'], axis=1, inplace=True)
```

```python
df_drop.head()
```

```python
df_drop['Genre'] = df_drop['Genre'].apply(lambda x: x.lstrip('[').rstrip(']'))
```

```python
df_drop['Genre']
```

```python
df_drop[['Song Genre', 'Song Genre 2']] = df_drop['Genre'].str.split(',', 1, expand=True)
```

```python
df_drop.drop(['Song Genre 2'], axis=1, inplace=True)
```

```python
df_drop['Song Genre'] = df_drop['Song Genre'].apply(lambda x: x.replace("'", ""))
```

```python
df_drop.head()
```

```python
df_drop.drop(['Genre'], axis=1, inplace=True)
```

```python
df_drop.head()
```

```python
df_drop.tail()
```

```python
df_clean = df_drop
```

```python
df_drop['Streams'] = df_drop['Streams'].apply(lambda x: x.replace(",", ""))
```

```python
df_drop['Artist Followers'] = df_drop['Artist Followers'].apply(lambda x: x.replace(",", ""))
```

```python
df_drop.head()
```

```python
df_drop.Streams = pd.to_numeric(df_drop.Streams, errors="coerce")
```

```python
df_drop['Artist Followers'] = pd.to_numeric(df_drop['Artist Followers'], errors="coerce")
```

```python
df_drop['Popularity'].replace(' ', np.nan, inplace=True)
```

```python
df_drop.dropna(subset=['Popularity'], inplace=True)
```

```python
df_drop['Loudness'] = df_drop['Loudness'].apply(lambda x: x.lstrip('-'))
```

```python
df_drop.head()
```

```python
df_drop['Popularity'] = df_drop['Popularity'].astype(float)
```

```python
df_drop['Danceability'] = df_drop['Danceability'].astype(float)
df_drop['Energy'] = df_drop['Energy'].astype(float)
df_drop['Loudness'] = df_drop['Loudness'].astype(float)
df_drop['Speechiness'] = df_drop['Speechiness'].astype(float)
df_drop['Acousticness'] = df_drop['Acousticness'].astype(float)
df_drop['Liveness'] = df_drop['Liveness'].astype(float)
df_drop['Tempo'] = df_drop['Tempo'].astype(float)
df_drop['Valence'] = df_drop['Valence'].astype(float)
```

```python
df_drop['Loudnes'] = df_drop['Loudness'] / 100
```

```python
df_drop.head()
```

```python
df_drop.drop(['Loudness'], axis=1, inplace=True)
```

```python
df_drop.dtypes
```

```python
df_cleaned = df_drop
```

# Data Visualization

```python
genre_type = pd.DataFrame(df_cleaned.groupby(df_clean['Song Genre']).sum().reset_index())
```

```python
genre_type
```

```python
plt.figure(figsize=(16,14))
sns.barplot(x = 'Song Genre' , y = 'Streams' , data = genre_type, palette ='plasma')
plt.xticks(rotation=90, size=7)
plt.show()
```

### Okay so we may not see it much clearly but we can see that 'Dance Pop' type of songs has the highest number of streams in Spotify followed by 'latin' and 'Pop'. We cannot clearly see which is the lowest as a lot of them are lowest.

```python
genre_type['Total energy'].plot(label='Energy flow', figsize=(12,10))
genre_type['Popularity'].plot(label='Popularity flow')
plt.legend()
plt.show()
```
